```python
import pandas as pd
import numpy as np
from statsmodels.tsa.filters.hp_filter import hpfilter
```

```python
data = pd.read_csv("HP_data_new.csv")
lambda_values = [1600, 5000, 10000, 20000]
```

```python
# Columns to apply HP filter on
columns_to_filter = ['GDP', 'PRIV_CONS', 'GROSS_FIXED_CAPITAL']

# Apply HP filter to each column for each lambda value and append the trend component as a new column
for column in columns_to_filter:
    for lamb in lambda_values:
        cycle, trend = hpfilter(data[column], lamb=lamb)
        # Append the trend component as a new column with a name indicating the column, component, and lambda value
        data[f'{column}_trend_{lamb}'] = trend

# Save the updated DataFrame to a new CSV file
data.to_csv("HP_filtered_new.csv", index=False)
data
```
